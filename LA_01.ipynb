### 箭头 - quiver
parameters:
- X,Y 1D or 2D[网格数据];箭头的起始位置
- U,V 1D or 2D;终点位置
- C 颜色
- angles:'xy'代表(x,y)->(x+u,y+v),'uv'代表(x,y)->(u,v)
- scale:缩放因子,箭头变短/变长,值越大箭头越短;设置为1表示不缩放
- unit:原始单位
    - 'xy':$\sqrt{x^2+y^2}$;
    - 'inches':以图像英寸为单位;
    - 'width' or 'height':以图像高度和宽度为单位
- scale_units:缩放单位与unit参数一样


```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
```

```python
plt.rcParams['figure.dpi'] = 450
```

```python
# 流图
w = 3
# j特殊用法,代表几个点而非步长
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
U= np.cos(X) * np.sin(Y)
V= X+Y
fig,ax=plt.subplots()
ax.streamplot(X, Y, U, V, linewidth=2, density=1.5)
ax.set_axis_off()
```

```python
fig,ax=plt.subplots()
Y=(3,4)
# scale缩放因子
ax.quiver(*Y,angles='xy',scale_units='xy',units='xy',scale=1)
ax.set_xlim(-1, 8)
ax.set_ylim(-1, 8)
```

```python
X, Y = np.meshgrid(np.arange(0, 2), np.arange(0, 2))
U = np.ones_like(X)
V = np.ones_like(Y)

plt.quiver(X, Y, U, V)
plt.show()
X,Y,U,V
```

```python

X, Y = np.meshgrid(np.arange(5), np.arange(5))
U = np.ones_like(X)
V = np.ones_like(Y)

plt.figure(figsize=(12, 5))

# 左图：不缩放，真实向量长度
plt.subplot(1, 2, 1)
# plt.title("真实向量长度")
plt.quiver(X, Y, U, V, scale=10, scale_units='height')
plt.gca().set_aspect('equal')

# 右图：统一箭头大小
plt.subplot(1, 2, 2)
# plt.title("统一箭头大小")
plt.quiver(X, Y, U, V, scale=10, units='inches', scale_units='inches')
plt.gca().set_aspect('equal')
# plt.tight_layout()
plt.show()
```

### quiverkey 添加箭头图示

```python
x, y = np.meshgrid(np.arange(5), np.arange(5))
u = np.ones_like(x)
v = np.ones_like(y)

fig, ax = plt.subplots()

# 主图箭头
Q = ax.quiver(x, y, u, v, scale=10)

# 添加箭头图例
ax.quiverkey(Q, X=0.9, Y=1.05, U=1,
             label='1 unit arrow', labelpos='W')

plt.title("Quiver with arrow legend")
plt.tight_layout()
plt.show()
```

### colors
- RGB or RGBA(red,green,blue,alpha)是范围[0,1]的元组
- hex  比如#0f0f0f
- hex shorthand 比如#abc as #aabbcc 
- 单个[0,1]的字符串小数点,表示灰度
- single character 比如 'b'->blue

> https://matplotlib.org/stable/users/explain/colors/colors.html

```python
fig,ax=plt.subplots(figsize=(2,2))
re=patches.Rectangle((0,0),0.1,0.1)
re.set_color('0.1')
ax.add_patch(re)

re1=patches.Rectangle((0.1,0),0.1,0.1)
re1.set_color('red')
ax.add_patch(re1)

re2=patches.Rectangle((0.2,0),0.1,0.1)
re2.set_color('#0f0f0f')
ax.add_patch(re2)

re3=patches.Rectangle((0.3,0),0.1,0.1)
re3.set_color('#abc')
ax.add_patch(re3)
ax.set_axis_off()
```
