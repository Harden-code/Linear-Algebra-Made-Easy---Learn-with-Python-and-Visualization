### 行列式

行列式只为方阵定义:
- 几何解释的唯一性:行列式作为“体积（或面积）缩放因子”的直观几何意义，在输入输出空间维度相同时（方阵）才最有意义和明确。
- 代数性质的匹配:诸如“行列式非零则矩阵可逆”、“行列式非零则线性方程组有唯一解”等关键代数性质，是与方阵的结构和行为紧密绑定的。

```python
# 方程的解,分母判别
x,y,a_1,a_2,b_1,b_2,c_1,c_2=sp.symbols('x,y,a_1,a_2,b_1,b_2,c_1,c_2')
expr1=a_1*x+b_1*y-c_1
expr2=a_2*x+b_2*y-c_2
sp.linsolve([expr1,expr2],(x,y))
```
